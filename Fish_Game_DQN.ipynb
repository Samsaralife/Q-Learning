```python
import pygame
from pygame import *
import sys
import random

BACKGROUND = (0,191,255)
FISH_COLOR = (255,246,143)
WHITE = (255,255,255)
DOOR_COLOR = (255,255,255)
BALL_COLOR = (255,0,0)

SCREEN_SIZE = [450,600]
BALL_SIZE = [10,10]
FISH_SIZE = [10,30]
DOOR_SIZE = [100,12]

fish_speed = [3,3]
class Game(object):
    def __init__(self):
        pygame.init()
        self.clock = pygame.time.Clock()
        self.screen = pygame.display.set_mode(SCREEN_SIZE)
        pygame.display.set_caption('Robot_Fish Play Football')
        self.fish = pygame.image.load('/Users/wangyingbo/Desktop/fish.png')
        self.ball = pygame.image.load('/Users/wangyingbo/Desktop/ball.png')
        
        self.fish_rect = self.fish.get_rect()
        self.ball_rect = self.ball.get_rect()
        #初始化左球门的位置
        
        self.left_door_x = 70
        self.left_door_y = 30
        self.left_door_pos = pygame.Rect(self.left_door_x,self.left_door_y,DOOR_SIZE[0],DOOR_SIZE[1])
        #初始化右球门的位置
        self.right_door_x = 280
        self.right_door_y = 30
        self.right_door_pos = pygame.Rect(self.right_door_x,self.right_door_y,DOOR_SIZE[0],DOOR_SIZE[1])
        
        #球的初始化位置
        self.ball_pos_x = SCREEN_SIZE[0]//2 - BALL_SIZE[0]/2
        self.ball_pos_y = SCREEN_SIZE[1]//2 - BALL_SIZE[1]/2
        
        #球移动的方向
        
        self.ball_dir_x = -1
        self.ball_dir_y = -1
        #获得球在屏幕中的位置
        self.ball_pos = pygame.Rect(self.ball_pos_x,self.ball_pos_y,self.ball_rect.left,self.ball_rect.right)
        
        #分数
        self.score = 0
        
        #鱼的初始化位置
        self.fish_pos_x = SCREEN_SIZE[0]//2 - FISH_SIZE[0]/2
        self.fish_pos_y = SCREEN_SIZE[1] - 120
        
        #鱼的运动方向
        self.fish_dir_x = -1
        self.fish_dir_y = -1
        
        #获取鱼在屏幕中的位置
        self.fish_pos = pygame.Rect(self.fish_pos_x,self.fish_pos_y,self.fish_rect.left,self.fish_rect.right)
        self.cur_fish_state = 'up'
    #定义鱼的动作
    def fish_move_up(self):
        
        self.fish_pos.top -= fish_speed[1]
        self.fish_rect = self.fish_rect.move(0,fish_speed[1])
        
        print('current ball position:'+str(self.ball_pos.bottom))
        if self.fish_pos.top <= self.ball_pos.bottom and (self.fish_pos.left<=self.ball_pos.left and self.fish_pos.right>=self.ball_pos.left):
            self.ball_pos.top -= fish_speed[1]
            self.cur_fish_state = 'up'
        #self.fish_pos.move_ip(self.fish_pos_x,self.fish_pos_y)
        print('action====>up, current fish_pos_y:' + str(self.fish_pos.top))
    def fish_move_down(self):
        self.fish_pos.bottom += fish_speed[1]  
        self.cur_fish_state = 'down'
        print('action====>down, current fish_pos_y:' + str(self.fish_pos.top))
        
    def fish_move_left(self):
        
        self.fish_pos.left -= fish_speed[0]
        self.cur_fish_state = 'left'
        print('action====>left, current fish_pos_x:' + str(self.fish_pos.left))
        #self.fish_pos.move_ip(self.fish_pos_x,self.fish_pos_y)
    def fish_move_right(self):
        
        self.fish_pos.right += fish_speed[0]
        self.cur_fish_state = 'right'
        print('action====>right, current fish_pos_x:' + str(self.fish_pos.right))
        #self.fish_pos.move_ip(self.fish_pos_x,self.fish_pos_y)
        
    def get_current_state(self):
        return self.cur_fish_state
    
    def run(self):
        while True:
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    pygame.quit()
                    sys.exit()
                elif event.type == pygame.KEYDOWN:
                    if event.key == K_UP:
                        self.fish_move_up()
                        self.cur_fish_state = 'up'
                        print(self.get_current_state())
                    elif event.key == K_DOWN:
                        self.fish_move_down()
                        self.cur_fish_state = 'down'
                        print(self.get_current_state())
                    elif event.key == K_RIGHT:
                        self.fish_move_right()
                        self.cur_fish_state = 'right'
                        print(self.get_current_state())
                    elif event.key == K_LEFT:
                        self.fish_move_left()
                        self.cur_fish_state = 'left'
                        print(self.get_current_state())
            x = random.randint(-50,50)
            y = random.randint(-50,50)
            #获取鱼在屏幕中的位置
            #self.fish_pos = pygame.Rect(self.fish_pos_x,self.fish_pos_y,self.fish_rect.left,self.fish_rect.right)
            
            #self.fish_pos.move(self.fish_pos_x,self.fish_pos_y)
            
            self.screen.fill(BACKGROUND)
            self.fish_pos.left = self.fish_pos.x
            self.fish_pos.top = self.fish_pos.y
            self.screen.blit(self.fish,self.fish_pos)
            
            #鱼的临界点
            if self.fish_pos.top<=0:
                self.fish_pos.top += fish_speed[1]
            elif self.fish_pos.bottom >= SCREEN_SIZE[1]:
                self.fish_pos.bottom = 10
            elif self.fish_pos.left <= 0:
                self.fish_pos.left += fish_speed[1]
            elif self.fish_pos.left >= (SCREEN_SIZE[0]-30):
                self.fish_pos.right -= fish_speed[1]
                
            #球的临界点
            if self.ball_pos.top <= 0:
                self.ball_pos.top = 0
            elif self.ball_pos.bottom >= (SCREEN_SIZE[1]- 10):
                self.ball_pos.bottom = (SCREEN_SIZE[1]-100)
            elif self.ball_pos.left <= 0:
                self.ball_pos.left += fish_speed[0]
            elif self.ball_pos.right >= SCREEN_SIZE[0]-20:
                self.ball_pos.right -= fish_speed[0]

            if self.ball_pos.bottom <=30 and self.ball_pos.left <= 170 and self.ball_pos.right <=280:
                self.score +=10
                print('currcent score:'+str(self.score))
            self.screen.blit(self.fish,(self.fish_pos_x,self.fish_pos_y))
            #绘制球门，鱼，球的图形
            pygame.draw.rect(self.screen,DOOR_COLOR,self.left_door_pos)
            pygame.draw.rect(self.screen,DOOR_COLOR,self.right_door_pos)
            self.screen.blit(self.ball,self.ball_pos)
            #pygame.draw.rect(self.screen,BALL_COLOR,self.ball_pos)
            
            pygame.display.update()
            self.clock.tick(60)
game = Game()
game.run()
```
