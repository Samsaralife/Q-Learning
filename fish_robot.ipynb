```python
import pygame
from pygame import *
import sys
import random
from math import *
BACKGROUND = (0,191,255)
FISH_COLOR = (255,246,143)
WHITE = (255,255,255)
DOOR_COLOR = (255,255,255)
BALL_COLOR = (255,0,0)

SCREEN_SIZE = [450,600]
BALL_SIZE = [10,10]
FISH_SIZE = [10,30]
DOOR_SIZE = [100,12]

fish_speed = [3,3]
class Game(object):
    def __init__(self):
        pygame.init()
        self.clock = pygame.time.Clock()
        self.screen = pygame.display.set_mode(SCREEN_SIZE)
        pygame.display.set_caption('Robot_Fish Play Football')
        self.fish = pygame.image.load('/Users/wangyingbo/Desktop/WANG/fish.png').convert_alpha()
        self.ball = pygame.image.load('/Users/wangyingbo/Desktop/WANG/ball.png').convert_alpha()
        
        #初始化鱼的位置
        self.fish_position=[(SCREEN_SIZE[0]//2 - FISH_SIZE[0]/2),(SCREEN_SIZE[1] - 120)]
        self.fish_width = self.fish.get_width()  #图片中鱼的宽度
        self.fish_height = self.fish.get_height() # 图片中鱼的高度
        
        #球的初始化位置
        #self.ball_position = [random.randrange(1,40)*10,random.randrange(1,50)*10]
        self.ball_position = [(SCREEN_SIZE[0]//2 - BALL_SIZE[0]/2),(SCREEN_SIZE[1]//2 - BALL_SIZE[1]/2)]
        self.ball_width = self.ball.get_width() #图片中球的宽度
        self.ball_height = self.ball.get_height() #图片中球的高度
        
        #初始化左球门的位置
        
        self.left_door_x = 70
        self.left_door_y = 30
        self.left_door_pos = pygame.Rect(self.left_door_x,self.left_door_y,DOOR_SIZE[0],DOOR_SIZE[1])
        
        #初始化右球门的位置
        self.right_door_x = 280
        self.right_door_y = 30
        self.right_door_pos = pygame.Rect(self.right_door_x,self.right_door_y,DOOR_SIZE[0],DOOR_SIZE[1])
        
        self.ball_in_door = False
        
        self.score = 0
        self.is_goal = False
        
        self.vilocity = 800
        self.time = 1/1000
        self.distant = 0
        
        self.A = ()
        self.B = ()
        self.C = ()
        
        self.angle = [0,90,180,270,360]
        self.fish_state = 'up'
        
    #定义鱼和球的运动
    def move_up(self):
        
        current_fish_state = self.get_fish_state()
        
        print('current_fish_state: '+ current_fish_state)
        if current_fish_state == 'left':
            self.fish = pygame.transform.rotate(self.fish,self.angle[0])
        elif current_fish_state == 'down':
            self.fish = pygame.transform.rotate(self.fish,self.angle[2])
        elif current_fish_state == 'right':
            self.fish = pygame.transform.rotate(self.fish,self.angle[1])
        elif current_fish_state == 'up':
            print('test=========>')
            self.fish_state = 'up'
        
        
        print('fish_state====>' + self.fish_state)
        self.fish_position[1] -= 10
        
        if self.fish_position[1] <= (self.ball_position[1] + self.ball_height):
            if self.fish_position[0] + self.fish_width/2 >= self.ball_position[0] and (self.fish_position[0] + self.fish_width/2)  <= (self.ball_position[0] + 
                                                                                            self.ball_width):
                self.ball_position[1] -=10
                #self.ball_position = [random.randrange(1,40)*10,random.randrange(1,50)*10]
                self.score+=10
       
        if self.ball_position[0] == 0:
            self.ball_in_door = True
            pygame.quit()
            
    def move_down(self):
        
        current_fish_state = self.get_fish_state()
        self.fish_state == 'down'
        print('current_fish_state_down: '+ current_fish_state)
        print('test=======state: ' + self.fish_state)
        if current_fish_state == 'down':
            self.fish = pygame.transform.rotate(self.fish,self.angle[0])
        elif current_fish_state == 'up':
            self.fish = pygame.transform.rotate(self.fish,45)
        elif current_fish_state == 'left':
            self.fish = pygame.transform.rotate(self.fish,self.angle[1])
        elif current_fish_state == 'right':
            self.fish = pygame.transform.rotate(self.fish,self.angle[3])
        
        
        
        print('fish_state====>' + self.fish_state)
        self.fish_position += 10
        
    def move_left(self):
        
        current_fish_state = self.get_fish_state()
        self.fish_state = 'left'
        print('current_fish_state_left: '+current_fish_state)
        if current_fish_state == 'left':
            pass
        elif current_fish_state == 'right':
            self.fish = pygame.transform.rotate(self.fish,self.angle[2])
        elif current_fish_state == 'up':
            self.fish = pygame.transform.rotate(self.fish,self.angle[1])
        elif current_fish_state == 'down':
            self.fish = pygame.transform.rotate(self.fish,self.angle[3])
        
        
        print('fish_state====>' + self.fish_state)
        self.fish_position[0] -=10
        
    def move_right(self):
        
        current_fish_state = self.get_fish_state()
        self.fish_state = 'right'
        print('current_fish_state: '+ current_fish_state)
        if current_fish_state == 'right':
            pass
        elif current_fish_state == 'left':
            self.fish = pygame.transform.rotate(self.fish,self.angle[2])
        elif current_fish_state == 'up':
            self.fish = pygame.transform.rotate(self.fish,self.angle[3])
        elif current_fish_state == 'down':
            self.fish = pygame.transform.rotate(self.fish,self.angle[1])
            
        
        print('fish_state====>' + self.fish_state)
        self.fish_position[0] += 10
    
    def get_fish_state(self):
        return self.fish_state
    
    def get_score(self):
        return self.score
        
    def run(self):
        
        while True:
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    pygame.quit()
                    sys.exit()
                elif event.type == pygame.KEYDOWN:
                    if event.key == K_UP:
                        self.move_up()
                        
                    elif event.key == K_DOWN:
                        self.fish_position[1] +=10
                        
                    elif event.key == K_RIGHT:
                        self.fish_position[0]+=10
                    elif event.key == K_LEFT:
                        self.fish_position[0] -=10
           
            
            #鱼的边界问题
            
            if self.fish_position[0] < 0:  #左
                self.fish_position[0] = 0
            elif self.fish_position[0] > (SCREEN_SIZE[0] - self.fish_width):  #右
                self.fish_position[0] = SCREEN_SIZE[0] - self.fish_width
            elif self.ball_position[1] == self.ball_height:
                self.ball_position[1] = self.ball_height
                
            elif self.fish_position[1] <0:  #上
                self.fish_position[1] = 0
            elif self.fish_position[1] > (SCREEN_SIZE[1] - self.fish_height): #下
                self.fish_position[1] = SCREEN_SIZE[1] - self.fish_height
           
            #球的边界问题
            if self.ball_position[0] < 0: #左
                self.ball_position[0] = 0
                self.fish_position[0] = self.ball_position[0] + self.ball_width
            elif self.ball_position[0] > (SCREEN_SIZE[0] - self.ball_width):#右
                self.ball_position[0] = SCREEN_SIZE[0] - self.ball_width
                
                
            elif self.ball_position[1] <0: #上
                self.ball_position[1] = 0
                if(self.ball_position[1]==0):
                    self.fish_position[1] == self.ball_height
            elif self.ball_position[1] > (SCREEN_SIZE[1] - self.fish_height): #下
                self.ball_position[1] = SCREEN_SIZE[1] - self.fish_height
            #判断球是否在球门内，如果是，则分时加10
            if self.ball_position[0]>= 170 and self.ball_position[0]<=280 and self.ball_position[1]<= 30:
                self.score = 10
                self.is_goal = True
            self.screen.fill(BACKGROUND)
            self.screen.blit(self.fish,self.fish_position)
            self.screen.blit(self.ball,self.ball_position)
            #绘制球门，鱼，球的图形
            pygame.draw.rect(self.screen,DOOR_COLOR,self.left_door_pos)
            pygame.draw.rect(self.screen,DOOR_COLOR,self.right_door_pos)
            
            pygame.display.update()
            self.clock.tick(60)
if __name__ == '__main__':
    game = Game()
    game.run()
    print('current-score'+str(game.get_score()))
```
